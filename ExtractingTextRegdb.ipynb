```python
import pandas as pd
```

```python
with open('Softberry.txt') as f:
    for line in f:
        if line.startswith(' QUERY:'):
            query = line.split(':', 1)[1].strip()
        if 'AC:' in line:
            ac = line.split('AC:')[1].split(':')[0].strip()
            print(query,ac)
```

```python
df = pd.read_csv('Softberry.csv')
df.columns = ['Gene', 'Element']
df['Count'] = 1
df = df.pivot_table(index='Gene', columns='Element', aggfunc=sum)
print(df)
```

```python
file_name = 'RegDB.xlsx'
```

```python
df.to_excel(file_name) 
print('DataFrame is written to Excel File successfully.')
```
